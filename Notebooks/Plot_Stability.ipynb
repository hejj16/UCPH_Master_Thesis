```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

```

```python


def PLOTS(path):
    
    with open(path, "rb") as f:
        (Classify_train_p, Classify_test_p, Classify_train_s, Classify_test_s, Rec_Acc, Rec_Acc_remove_gap)  = pickle.load(f)
    
    plt.rcParams["figure.figsize"] = (5, 5)
    plt.subplot(2, 1, 1)
    for i in Classify_test_p:
        plt.ylim(0.3, 0.75)
        plt.plot( np.arange(1, 21, 1), i[:20])
    plt.ylim(0.3, 0.75)
    plt.yticks(np.arange(0.3, 0.75, 0.05))
    plt.xticks(np.arange(1, 21, 1))
    plt.grid()
    plt.ylabel("Pearson")

    plt.subplot(2, 1, 2)
    for i in Classify_test_s:
        plt.ylim(0.3, 0.75)
        plt.plot( np.arange(1, 21, 1), i[:20])
    plt.ylim(0.3, 0.75)
    plt.yticks(np.arange(0.3, 0.75, 0.05))
    plt.xticks(np.arange(1, 21, 1))
    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Spearman")
    plt.show()
    plt.show()



    plt.rcParams["figure.figsize"] = (5, 5)
    plt.subplot(2, 1, 1)
    plt.plot( np.arange(1, 21, 1), np.array(Classify_test_p).mean(0)[:20], color="tab:orange", linewidth=2)

    plt.fill_between(np.arange(1, 21, 1), 
                     np.array(Classify_test_p).mean(0)[:20] + np.array(Classify_test_p).std(0)[:20], 
                     np.array(Classify_test_p).mean(0)[:20] - np.array(Classify_test_p).std(0)[:20],
                     facecolor='gray', 
                    alpha=0.3) 
    plt.scatter(np.arange(1, 21, 1), np.array(Classify_test_p).max(0)[:20], marker="x", color="black", s=15 )
    plt.scatter(np.arange(1, 21, 1), np.array(Classify_test_p).min(0)[:20], marker="x", color="black", s=15 )

    plt.ylim(0.3, 0.75)
    plt.yticks(np.arange(0.3, 0.75, 0.05))
    plt.xticks(np.arange(1, 21, 1))
    plt.grid()
    plt.ylabel("Pearson")
    
    print("Pearson: ")
    print(np.array(Classify_test_p).mean(0)[19])
    print(np.array(Classify_test_p).std(0)[19])


    plt.subplot(2, 1, 2)
    plt.plot(np.arange(1, 21, 1), np.array(Classify_test_s).mean(0)[:20], color="tab:orange", linewidth=2)

    plt.fill_between(np.arange(1, 21, 1), 
                     np.array(Classify_test_s).mean(0)[:20] + np.array(Classify_test_s).std(0)[:20], 
                     np.array(Classify_test_s).mean(0)[:20] - np.array(Classify_test_s).std(0)[:20],
                     facecolor='gray', 
                    alpha=0.3) 
    plt.scatter(np.arange(1, 21, 1), np.array(Classify_test_s).max(0)[:20], marker="x", color="black", s=15 )
    plt.scatter(np.arange(1, 21, 1), np.array(Classify_test_s).min(0)[:20], marker="x", color="black", s=15 )

    plt.ylim(0.3, 0.75)
    plt.yticks(np.arange(0.3, 0.75, 0.05))
    plt.xticks(np.arange(1, 21, 1))
    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Spearman")
    plt.show()
    
    print("Spearman: ")
    print(np.array(Classify_test_s).mean(0)[19])
    print(np.array(Classify_test_s).std(0)[19])
```
